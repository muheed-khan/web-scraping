Scraping Yahoo Finance for Practice

```python
import requests
url = "https://finance.yahoo.com/"
html = requests.get(url).content
```

```python
from scrapy import Selector
sel = Selector(text=html)
```

```python
cssLocator = 'span.symbol.valid::text'
companiesName = sel.css(cssLocator).getall()
companiesName
```

```python
cssLocator = 'a.none-link.fin-size-small.tw-block::attr(href)'
links = sel.css(cssLocator).getall()
links
```

```python
cssLocator = 'fin-streamer.last-price > span::text'
price = sel.css(cssLocator).getall()
price
```

```python
cssLocator = 'fin-streamer.percentChange > span::text'
change_in_percentage = sel.css(cssLocator).getall()
change_in_percentage
```

```python
positive_changed_stocks = [
    (name, link, prices, change) 
    for name, link, prices, change in zip(companiesName, links, price, change_in_percentage)
    if change.startswith('+')
]
```

```python
positive_changed_stocks
```

```python
import pandas as pd
df = pd.DataFrame(positive_changed_stocks, columns=['Company Title','Link','Price','Change in %'])
df
```
